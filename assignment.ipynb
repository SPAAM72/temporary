```python
raw_data1 = pd.read_csv('internshala.csv')
#import of cleaned data for clustering ie problem statement number 3
```

```python
#display(raw_data1)
```

```python
#making small changes
raw_data1  = raw_data1.applymap(lambda x: 1 if x == True else x)
raw_data1 =  raw_data1.applymap(lambda x: 0 if x == False else x)
```

```python
#display(raw_data1)
```

```python
#deleting redundant rows
raw_data1.drop(['maximumMarks','minusMark'],axis=1)
```

```python
info = raw_data1[['_id','studentName']]
```

```python
data = raw_data1.drop(['_id','studentName'],axis = 1)
```

```python
#display(data)
```

```python
data.corr()
```

```python
import numpy as np
```

```python
#normalising the data
data[['offscreenTime','totalTime']] = np.log(data[['offscreenTime','totalTime']])
```

```python
data.head()
```

```python
import math
```

```python
#more data cleaning

data = data.replace([np.inf,-np.inf],1.0)
data = data.replace(np.nan,0.0)
```

```python
#applying dimensionality reduction
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(data)

reduced_data = pca.transform(data)
reduced_data = pd.DataFrame(reduced_data, columns = ['Dimension 1', 'Dimension 2'])
```

```python
#applying possible clusters among students and calculating silhouette score using Kmeans

from sklearn.cluster import KMeans

for i in reversed(range(2,10)):
    clusterer = KMeans(n_clusters=i ,random_state=42).fit(reduced_data)
    
    # Predict the cluster for each data point
    preds = clusterer.predict(reduced_data)

    # Find the cluster centers
    centers = clusterer.cluster_centers_
 
    # Calculate the mean silhouette coefficient for the number of clusters chosen
    from sklearn.metrics import silhouette_score
    score = silhouette_score(reduced_data, preds)

    print ("Score for {} clusters = {}  ".format(i,score))
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import pandas as pd
import numpy as np
def cluster_results(reduced_data, preds, centers):
	'''
	Visualizes the PCA-reduced cluster data in two dimensions
	Adds cues for cluster centers 
	'''

	predictions = pd.DataFrame(preds, columns = ['Cluster'])
	plot_data = pd.concat([predictions, reduced_data], axis = 1)

	# Generate the cluster plot
	fig, ax = plt.subplots(figsize = (14,8))

	# Color map
	cmap = cm.get_cmap('gist_rainbow')

	# Color the points based on assigned cluster
	for i, cluster in plot_data.groupby('Cluster'):   
	    cluster.plot(ax = ax, kind = 'scatter', x = 'Dimension 1', y = 'Dimension 2', \
	                 color = cmap((i)*1.0/(len(centers)-1)), label = 'Cluster %i'%(i), s=30);

	# Plot centers with indicators
	for i, c in enumerate(centers):
	    ax.scatter(x = c[0], y = c[1], color = 'white', edgecolors = 'black', \
	               alpha = 1, linewidth = 2, marker = 'o', s=200);
	    ax.scatter(x = c[0], y = c[1], marker='$%d$'%(i), alpha = 1, s=100);

	

	# Set plot title
	ax.set_title("PLOT by KMeans");
```

```python
cluster_results(reduced_data, preds, centers)
```

```python
#kmeans has not done well based on the above visualization DBSCAN is the best algorithm
from sklearn.cluster import DBSCAN

clustering = DBSCAN(eps=4, min_samples=10).fit(reduced_data)
preds1 = clustering.labels_
```
